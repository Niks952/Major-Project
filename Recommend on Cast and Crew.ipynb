```python
data=pd.read_csv('movies_cleaned_data.csv')
```

```python
data['tags']=data['crew']+data['cast']
```

```python
movie_genre=data[['movie_id','title','tags']]
```

```python
movie_genre.head()
```

Stemming

```python
import nltk
from nltk.stem.porter import PorterStemmer
```

```python
ps=PorterStemmer()
```

```python
def stem(text):
    L=[]
    for i in text.split():
        L.append(ps.stem(i))
    return " ".join(L)
```

```python
movie_genre['tags']=movie_genre['tags'].apply(stem)
```

End Stemming

```python
from sklearn.feature_extraction.text import TfidfVectorizer
cv = TfidfVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(movie_genre['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_credits=cosine_similarity(vectors)
```

```python
similarity_credits
```

```python
def recommend_credits(movie):
    movies_index=movie_genre[movie_genre['title']==movie].index[0]
    distances=similarity_credits[movies_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:21]
    
    for i in movies_list:
        print(movie_genre.iloc[i[0]].title)
    
    return
```

```python
recommend_credits('Avatar')
```
